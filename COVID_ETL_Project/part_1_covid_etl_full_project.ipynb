<h4> 1.5 Download the CSV files in S3 anthena_query_output Folder and Load into pandas DataFrames

```python
# This function is to 
# 1) retrieve the csv file of a previously executed query in S3 athena_query_output folder using QueryExecutionId, 
# 2) download the results to a temporary file location on local computer,
# 3) load the results into a pandas DataFrame.
Dict = {}
def download_and_load_query_results(client: boto3.client, query_response: Dict) -> pd.DataFrame:
    # This while loop is to ensure the download_file function can be executed without any error caused by the 'query_response' input.
    while True:
        # Try to get the first 1000 rows of the query results
        try:
            # get_query_results method is part of the Athena client in boto3 library
            # retrieves the results of a query that was executed previously in the 'query_response' input
            client.get_query_results(QueryExecutionId = query_response['QueryExecutionId'])
            break
        # If an exception occurs
        except Exception as err:
            # If the exception is because the query has not yet finished, 
            if 'not yet finished' in str(err):
                # wait for 0.001s and then go back and start 'try' again.
                time.sleep(0.001)
            # If the exception is due to other reasons,
            else:
                # raise the exception: the error will propagate up and the function will halt.
                raise err

    # Initialize a new S3 client to use S3 services
    s3_client = boto3.client(
        's3',
        aws_access_key_id = aws_access_key,
        aws_secret_access_key = aws_secret_key,
        region_name = aws_region)

    # Define the local temporary file name for storing the downloaded query result
    temp_file_location: str = 'athena_query_results.csv'
    # Use the S3 client to download the query result to the temporary file
    # When the function is runned with new response inputs, the content in athena_query_results.csv will also be updated.
    s3_client.download_file(
        s3_raw_data_dir_bucket,
        # The result of the query executed previously in the 'query_response' input
        f"{s3_raw_data_dir_folder}/{query_response['QueryExecutionId']}.csv", 
        temp_file_location)

    # Load the downloaded query result into a pandas DataFrame and return the DataFrame.
    return pd.read_csv(temp_file_location)
```
